# Busca - Problema das n-rainhas

F = Número de rainhas que não se atacam <br>
Mínimo: 0 (meta) <br>
Máximo: Comb (N,2) = N! / ((N-2)!*2!)-N*(N-1)/2 <br>
Para N = 8  <br>
MAX = 8! / ((8-2)!*2!)-8*7/2 = 28

```python
# Função para gerar um tabuleiro
def geraTabuleiro(tamanho):
    if tamanho < 2 : return [-1]
    tab = []
    for i in range(tamanho):
        l = []
        for j in range(tamanho):
            l.append(0)
        tab.append(l)
    for i in range(tamanho): tab[i][0] =1
    return tab
```

```python
# Função para imprimir o tabuleiro
def printTabuleiro(tab):
    for i in tab: print(i)
```

```python
# Função para localizar as rainhas
def locateQueens(estado):
    q = []
    for i in range(len(estado)):
        for j in range(len(estado)):
            if estado[i][j] == 1: q.append((i,j))
    return q
```

```python
# Função para calcular a quantidade de ataques no tabuleiro
def calAtaques(estado):
    at = 0
    # ataques em linha
    atlinhas = 0
    for i in estado:
        if sum(i) > 1:
            atlinhas += sum(i)-1
    at += atlinhas
    
    # Ataque em colunas
    atcolunas = 0
    for c in range(len(estado)):
        nl = 0
        for l in range(len(estado)): nl += estado[l][c]
        if nl>1: atcolunas += nl -1
    at += atcolunas
    
    # Ataques na diagonal
    pr = locateQueens(estado)
    d=0
    
    for p, coords in enumerate(pr):
        i,j = coords
        # Diagonal esquerda
        k,w = i+1, j-1
        while(w>=0 and k<len(estado)):
            if estado[k][w] == 1:
                d+=1
                break
            w -= 1
            k += 1
        # Diagonal Direita
        l,c = i+1, j+1
        while(l < len(estado) and c< len(estado)):
            if estado[l][c] ==1:
                d+=1
                break
            l += 1
            c += 1
    
    at += d
    return at
    
```

```python
# Função para movimentar uma rainha
def moveRainhaAleatoria(estado, passo):
    tam = len(estado)
    while True:
        i,j = random.choice(locateQueens(estado))
        k = (j + passo)%tam
        if estado[i][k] == 0:
            estado[i][k], estado[i][j] = estado[i][j], estado[i][k]
            break
    return estado
            
```

```python
def buscaCega(estado_Atual, MAX):
#   n = len(estado[0])
  ataques = []
  i = 0

  estado_Anterior = copy.deepcopy(estado_Atual)

  while i < MAX:
    atqAnterior = calAtaques(estado_Anterior)

    estado_Atual = moveRainhaAleatoria(copy.deepcopy(estado_Anterior), random.randint(1, 5))

    atqAtual = calAtaques(estado_Atual)
    
    if atqAtual <= atqAnterior:
      ataques.append(atqAtual)
      estado_Anterior = copy.deepcopy(estado_Atual)

    if atqAtual == 0 : break
      
    i += 1
    

  # Gráfico da Busca
  x = range(len(ataques))
  plt.plot(x, ataques)
  plt.xlabel("Rodadas")
  plt.ylabel("Ataques")
  plt.show()

  print("Rodada: ", i)

  printTabuleiro(estado_Atual)
```

```python
def buscaHC(estado, passo, MAX):
    c = 0
    n = len(estado[0])
    ataques = []
    change = True

    while c < MAX:
        change = False
        atq = calAtaques(estado)
        ataques.append(atq)

        if atq == 0:
            return "Solução", estado, atq, ataques

        suc = moveRainhaAleatoria(copy.deepcopy(estado), passo)
        
        if calAtaques(suc) <= atq:
            estado = suc
            change = True

        if not change: c += 1

    return "Falha", estado, atq, ataques
```

```python
def gera_vizinho(estado_atual):
    """
    Gera um estado vizinho aleatório para o problema das N-Rainhas.
    """
    # Seleciona uma coluna aleatória.
    coluna = random.randint(0, len(estado_atual) - 1)
    # Seleciona uma linha aleatória diferente da atual.
    linha = random.randint(0, len(estado_atual) - 1)
    while linha == estado_atual[coluna]:
        linha = random.randint(0, len(estado_atual) - 1)
    # Cria um novo estado com a nova posição da rainha.
    novo_estado = list(estado_atual)
    novo_estado[coluna] = linha
    return tuple(novo_estado)
```

```python
def perturb(state):
    n = len(state)
    # Seleciona uma posição aleatória para modificar
    i = random.randint(0, n-1)
    j = random.randint(0, n-1)
    # Gera um novo estado com a posição modificada
    new_state = list(state)
    new_state[i] = j
    return tuple(new_state)

def get_neighbors(state, tamanho):
    neighbors = []
    visited = []
    while len(neighbors) < tamanho:
        neighbor = perturb(state)
        if neighbor not in visited:
            visited.append(neighbor)
            neighbors.append(neighbor)
    return neighbors
```

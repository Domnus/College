```python
from sklearn import datasets
dados = datasets.load_iris()
print(dados)
```

```python
print(dados.target_names)
print(dados.target)
```

```python
print(dados.feature_names)
print(dados.data)
```

```python
import pandas as pd
total = pd.DataFrame(dados.data, columns=dados.feature_names)
total['target'] = dados.target
total.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plot = sns.FacetGrid(total, hue='target')
plot.map(sns.distplot, "sepal length (cm)").add_legend()

plot = sns.FacetGrid(total, hue='target')
plot.map(sns.distplot, "sepal width (cm)").add_legend()

plot = sns.FacetGrid(total, hue='target')
plot.map(sns.distplot, "petal length (cm)").add_legend()

plot = sns.FacetGrid(total, hue='target')
plot.map(sns.distplot, "petal width (cm)").add_legend()

plt.show()
```

```python
X = dados.data[:, 2:4]
print(X)
```

```python
parcial = pd.DataFrame(data=X, columns=dados.feature_names[2:4])
parcial['target'] = dados.target
parcial.head()
```

```python
import matplotlib.pyplot as plt
parcial.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)', c='target', cmap='viridis')
```

```python
import seaborn as sns
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=parcial, palette='Set1')
```

```python
from sklearn.model_selection import train_test_split
y = dados.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print("Total de registros com características para treino: ", X_train.shape)
print("Total de registros com rótulos para treino........: ", y_train.shape)
print("Total de registros com características para teste.: ", X_test.shape)
print("Total de registros com rótulos para teste.........: ", y_test.shape)
```

```python
from sklearn import tree 
modelo = tree.DecisionTreeClassifier()
modelo.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 12))

grafico = tree.plot_tree(modelo, feature_names = dados.feature_names,
                         class_names = list(dados.target_names), rounded = True, 
                         filled = True, fontsize=14)
```

```python
import numpy as np
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = modelo.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel('petal length (cm)'); plt.ylabel('petal width (cm)')
```

```python
previsao = modelo.predict(X_test)
print(previsao)
```

```python
previstos = pd.DataFrame(data=X_test, columns=dados.feature_names[2:4])
previstos['target'] = previsao
previstos.head()
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("Relatório de classificação: ")
print(classification_report(y_test, previsao))
print("Acurácia do modelo: ", accuracy_score(y_test, previsao))
```

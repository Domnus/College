```python
from sklearn import datasets
dados = datasets.load_iris()
print(dados)
```

```python
print(dados.target_names)
```

```python
print(dados.target)
```

```python
print(dados.feature_names)
```

```python
print(dados.data)
```

```python
import pandas as pd 
total = pd.DataFrame(data=dados.data, columns = dados.feature_names)
total.head()
```

```python
from sklearn.preprocessing import normalize
X = dados.data
y = dados.target
Xnorm = normalize(X)
print(Xnorm)
```

```python
from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=3, init='random')
modelo.fit(Xnorm)
```

```python
print(modelo.cluster_centers_)
```

```python
distancia = modelo.fit_transform(Xnorm)
print(distancia)
```

```python
print(y)
```

```python
rotulos = modelo.labels_
print(rotulos)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Xnorm[:,0], Xnorm[:,3], s=50)
plt.title('Dataset Iris Sem Agrupamento')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(Xnorm[:,0], Xnorm[:,3], s=50, c=rotulos)
plt.title('Dataset Iris Sem Agrupamento')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(Xnorm[:, 0], Xnorm[:, 3], s=50, c=rotulos)
plt.scatter(modelo.cluster_centers_[:,0], modelo.cluster_centers_[:,3], s=200, c='red', label='centroids')
plt.title('Dataset Iris com Agrupamento e os Centroides')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
